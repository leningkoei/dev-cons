```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel("exercise5/操作票.xls", header = 4)
data
```

```python
mask1 = data["操作命令内容"].str.contains("拉开|合上", na = False)
mask1
```

```python
mask2 = data["操作命令内容"].str.contains("站内", na = False)
mask2
```

```python
data2= data[mask1 & mask2]
data2
```

```python
data3 = data2[["序号", "下令时间", "操作命令内容"]]
data3
```

```python
# warning: data2["操作命令内容"] = data2["操作命令内容"].str.replace("站内", "")
# data2.loc[:, "操作命令内容"] = data2["操作命令内容"].str.replace("站内", "")
data4 = data3.copy()
data4["操作命令内容"] = data4["操作命令内容"].str.replace("站内", "")
data4
```

```python
data4["开合操作"] = data4["操作命令内容"].str.extract(r'(拉开|合上)')
data4
```

```python
data4["开关"] = data4["操作命令内容"].str.extract(r'(拉开|合上)(.*负\d+?)')[1]
data4
```

```python
data5 = data4.sort_values(by = ["下令时间", "序号"])
data5
```

```python
data6 = data5.drop_duplicates("开关", keep = "last")
data6
```

```python
data7 = data6.copy()
data7["分合闸"] = data7["开合操作"].map(lambda x : "分闸" if x == "拉开" else "合闸")
data7
```
