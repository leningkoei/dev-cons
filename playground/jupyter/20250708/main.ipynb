```python
import pandas as pd
```

## 表1 操作票

```python
# 以操作票.xls中第5行（即序号、计划操作时间...所在行）为标题行
df1 = pd.read_excel('data/操作票.xls', header=4)
df1
```

```python
mask1 = df1['操作命令内容'].str.contains('拉开|合上', na=False)
mask1
```

```python
mask2 = df1['操作命令内容'].str.contains('站内', na=False)
mask2
```

```python
mask1 & mask2
```

```python
df1_1 = df1[mask1 & mask2]
df1_1
```

```python
df1_2 = df1_1[['序号', '下令时间', '操作命令内容']]
df1_2
```

```python
# 为了便于提取，去掉“站内”2个字
df1_2.loc[:, '操作命令内容'] = df1_2['操作命令内容'].str.replace('站内', '')
df1_2
```

```python
df1_2.info()
```

```python
# 下令时间转化为时间格式
df1_2.loc[:, '下令时间'] = pd.to_datetime(df1_2['下令时间'])
df1_2['下令时间']
```

```python
df1_2 = df1_2.copy()
df1_2['开合操作'] = df1_2['操作命令内容'].str.extract(r'(拉开|合上)')
df1_2['开合操作']
```

```python
df1_2['操作命令内容'].str.extract(r'(拉开|合上)(.*?负\d+?)')
```

```python
df1_2['开关'] = df1_2['操作命令内容'].str.extract(r'(拉开|合上)(.*?负\d+?)')[1]
df1_2
```

```python
# 按下令时间升序排列，时间相同的话按序号排列
df1_3 = df1_2.sort_values(by=['下令时间', '序号']).copy()
df1_3
```

```python
df1_4 = df1_3.drop_duplicates('开关', keep='last')
df1_4
```

```python
df1_5 = df1_4.copy()
df1_5['分合闸'] = df1_5['开合操作'].apply(lambda x: '分闸' if x == '拉开' else '合闸')
df1_5
```
